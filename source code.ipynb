```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import time

# get the start time
st = time.time()
```

```python
# opening DNA_Motifs.txt the file in read mode
my_file = open(r"C:\Users\Lenovo\Desktop\freelancing data\Take_Home_Assignment_2\DNA_Motifs.txt", "r")
# reading the file
data = my_file.read()
# converting DNA_Motifs.txt file to list of motifs
DNA_Motifs = data.replace('\n', ' ').split(" ")
# printing the data
# print(DNA_Motifs)
```

```python
# opening DNA_Seqs.txt the file in read mode
my_file = open(r"C:\Users\Lenovo\Desktop\freelancing data\Take_Home_Assignment_2\DNA_Seqs.txt", "r")
# reading the file
data = my_file.read()
# convert DNA_Seqs.txt sequences into a list
DNA_Seqs = data.replace('\n', ' ').split(" ")
# printing the data
# print(DNA_Seqs)
```

```python
# making a function which will find motifs in a dna sequence,pass two variable motifs and sequence to the motif finder
def motif_finder(motifs,sequence):
    # open a file named DNA_Seq_Analysis_Results.txt in a write mode from specified directory in order to save the resultant files in it
    with open(r'C:\Users\Lenovo\Desktop\freelancing data\Take_Home_Assignment_2\DNA_Seq_Analysis_Results.txt', 'w') as f:
        # make outer looop which will pass through every sequence given in sequence file
        for i in range(len(sequence)-1):
            # now open that saved DNA_Seq_Analysis_Results.txt file in append mode in ordr to add the new lines to the file
            with open(r'C:\Users\Lenovo\Desktop\freelancing data\Take_Home_Assignment_2\DNA_Seq_Analysis_Results.txt', 'a') as f:
                # f.write function writes txt in the specified txt file.
                # this line of code will add a line like "sequence 1 : ANDKXDJKNDKDJE",
                # as i will start from 0,so here i will be 0 and sequence[i] wil print the sequence wich is at zeroth location of seq list.
                f.writelines(('Sequence {} : {}'.format(i,sequence[i]),'\n'))
                # printing the length of seq present at sequence[i] location.
                f.writelines(('length of sequence = {}'.format(len(sequence[i])),'\n'))
            # making inner loop to pass through every motif in motif list
            # this line will work like this:
            # i=0,j=0 so when loop run for the first time , parameters for comparison will be the 1st seq and 1st motif.
            # in 2nd cycle i=0,j=1. now seq will be 1st,but motif will be 2nd. the cycle continue untill all the motifs are compard with 1st seq. 
            for j in range(len(motifs)):
                # again open the file as append mode to add the lines from inner loop to txt file.
                with open(r'C:\Users\Lenovo\Desktop\freelancing data\Take_Home_Assignment_2\DNA_Seq_Analysis_Results.txt', 'a') as f:
                    # the count(motifs[j]) counts the number of occurances of a motif in a seq.
                    # this line prints like "motif location_of_motif:motif_name:occurance_in_seq"
                    # for example motif 0 : AA : 3
                    f.write(('Motif {} : {} : {}\n'.format(j,motifs[j],str(sequence[i].count(motifs[j])))))
            with open(r'C:\Users\Lenovo\Desktop\freelancing data\Take_Home_Assignment_2\DNA_Seq_Analysis_Results.txt', 'a') as f:
                f.writelines(('=================================================================================================\n')) 
                       
```

```python
motif_finder(DNA_Motifs,DNA_Seqs)
```

```python
# define counters for diff motifs
AA_count=0
GA_count=0
CG_count=0
TT_count=0
GT_count=0
TAC_count=0
GGA_count=0
CTG_count=0
TAC_count=0
CATG_count=0
TAGT_count=0
TGATC_count=0
TTAAC_count=0
GATACG_count=0
TTGCGC_count=0
AGGCCGAC_count=0
# generate a loop to search all the motifs in earch seq. outer loop searchs seq and inner loop search mtifs in seq using inner loop
for i in range(len(DNA_Seqs)-1):
    # inner loop for motifs
    for j in range(len(DNA_Motifs)):
        # add 1 to a motif counter if a specific motif is present in a seq.
        # this will work like this
        # 1st cycle i=0,j=0
        # for example motif at i=0 is AA and seq at j=0 is CMKJDMAAKDAA
        # chck if AA==AA. count this occurance in seq 0,this cycle contue untill it count motif AA in all the sequences given in the file.
        if DNA_Motifs[j]=='AA':
                 AA_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='GA':
                 GA_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='CG':
                 CG_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='TT':
                 TT_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='GT':
                 GT_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='TAC':
                 TAC_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='GGA':
                 GGA_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='CTG':
                 CTG_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='CATG':
                 CATG_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='TAGT':
                 TAGT_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='TTAAC':
                 TTAAC_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='GATACG':
                 GATACG_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='TTGCGC':
                 TTGCGC_count+=DNA_Seqs[i].count(DNA_Motifs[j])
        if DNA_Motifs[j]=='AGGCCGAC':
                 AGGCCGAC_count+=DNA_Seqs[i].count(DNA_Motifs[j])
# save occurances of al the motifs in a list
COUNT_LIST=[AA_count,GA_count,CG_count,TT_count,GT_count,TAC_count,GGA_count,CTG_count,TAC_count,CATG_count,TAGT_count,
            TGATC_count,TTAAC_count,GATACG_count,TTGCGC_count,AGGCCGAC_count]
# print(COUNT_LIST)
# fin max and min number of occurances of the motifs.
most_frequent=max(COUNT_LIST)
less_frequent=min(COUNT_LIST)
# i have made a dictionary in which i save the motifs and its respective number of occurances in all the sequences.
res = {}
for key in DNA_Motifs:
    for value in COUNT_LIST:
        res[key] = value
        break
# find which motif has highest and lowest value
most_freq_motif = max(res, key=res.get)
less_freq_motif = min(res, key=res.get)
# writing the maximum and min frequency of occurances of motifs in sequences file.
with open(r'C:\Users\Lenovo\Desktop\freelancing data\Take_Home_Assignment_2\DNA_Seq_Analysis_Results.txt', 'a') as f:
        f.writelines(('MAximun frequency of {} for motif (s) {} \n'.format(most_frequent,most_freq_motif)))
        f.writelines(('MAximun frequency of {} for motif (s) {} \n'.format(less_frequent,less_freq_motif)))
```
